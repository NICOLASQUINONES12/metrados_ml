```python
df["Especialidad"] = (
    df["Especialidad"]
    .str.strip()      # elimina espacios
    .str.lower()      # todo en minúsculas
)
df["Especialidad"].unique()
```

```python
df_encoded = pd.get_dummies(df, columns=["Especialidad"])
df_encoded
```

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

loo = LeaveOneOut()
model = LinearRegression()

X = df_encoded[['Especialidad_estructuras','Especialidad_encofrados','Especialidad_acabados','Especialidad_arquitectura']]
y = df_encoded["Duración del proyecto"]

r2_scores = []

for train_index, test_index in loo.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2_scores.append(r2_score(y_test, y_pred))

np.mean(r2_scores)
```
